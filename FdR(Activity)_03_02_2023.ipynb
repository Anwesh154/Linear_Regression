## DataSet2_Linear Regression

```python
df=pd.read_csv("C:/Users/dell/Downloads/Flipkart Mobile - 2.csv")
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
plt.scatter(df['sales_price'],df['display_size'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df[['ROM','RAM','display_size','num_rear_camera','num_front_camera','discount_percent','battery_capacity','ratings','sales']]
y=df['sales_price']

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
lg=LinearRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
lg.score(X_test,y_test)
```

```python
lg.coef_
```
