```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Load the data
df = pd.read_csv('C:/Users/dell/Downloads/Kmeans data.csv')

# Select the features to cluster on
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply the k-means algorithm with k=5
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clusters
plt.scatter(X[y_kmeans == 0]['Age'], X[y_kmeans == 0]['Spending Score (1-100)'], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1]['Age'], X[y_kmeans == 1]['Spending Score (1-100)'], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2]['Age'], X[y_kmeans == 2]['Spending Score (1-100)'], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3]['Age'], X[y_kmeans == 3]['Spending Score (1-100)'], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4]['Age'], X[y_kmeans == 4]['Spending Score (1-100)'], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 2], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Load the data
df = pd.read_csv('C:/Users/dell/Downloads/Kmeans data.csv')
df=df.set_index('CustomerID')
df['Genre']=df['Genre'].replace({'Female':1,'Male':0})

df=df.rename(columns={'Genre':'Gender'})

X = df['Age']
Y = df['Spending Score (1-100)']
plt.scatter(X,Y)

plt.show()

data=list(zip(X,Y))
inertias=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1,11),inertias,marker='o')
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()

# Initialize the KMeans object with 2 clusters
kmeans = KMeans(n_clusters=4)

# Fit the KMeans model to the data
kmeans.fit(df[['Age','Spending Score (1-100)']])

# Get the labels of the data points
labels = kmeans.labels_


# Get the centroids of the clusters
centroids = kmeans.cluster_centers_

# Example scatter plot of the first two features with points colored according to cluster
plt.scatter(df['Age'], df['Spending Score (1-100)'], c=labels)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
